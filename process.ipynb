```python
aqi_history = []
for location in locations:
    # print(location)
    data = requests.get(url=f'http://air4thai.com/forweb/getHistoryData.php?stationID={location}&param=PM25&type=hr&sdate=2023-10-01&edate=2025-07-24&stime=00&etime=23')
    output = data.json()
    # print(len(output))
    stations = output['stations']
    assert len(stations) == 1, len(stations)
    # print(len(stations[0]['data']))
    with open(f'{location}.json', 'w') as f:
        json.dump(stations[0], f)
    
    station_id = stations[0]["stationID"]
    data_with_added_station_id = [
        {
            "DATETIMEDATA": datum["DATETIMEDATA"],
            "PM25": datum["PM25"],
            "stationid": station_id
        }
        for datum
        in stations[0]['data']
    ]
    aqi_history.extend(data_with_added_station_id)

df_aqi_history = pd.DataFrame.from_records(aqi_history)
```

```python
df_aqi_history = df_aqi_history.rename(columns={'stationid': 'stationID'})
```

```python
df_aqi_history_with_locations = df_aqi_history.set_index('stationID').join(df_locations.set_index('stationID'), on='stationID').reset_index()
df_aqi_history_with_locations
```

```python
gdf_aqi_history_with_locations = gpd.GeoDataFrame(df_aqi_history_with_locations, geometry=gpd.GeoSeries.from_xy(df_aqi_history_with_locations['long'], df_aqi_history_with_locations['lat'], crs='EPSG:4326'))
```
